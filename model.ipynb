```python
#import necessary
import pandas as pd
from sklearn.linear_model import LinearRegression #so we can use linear regression model
from sklearn.model_selection import train_test_split #to split dataset into train and test
from sklearn.metrics import mean_squared_error, r2_score #mse and r2 to evaluate our predictions
```

```python
#load dataset
df = pd.read_csv('cleaned2.csv')
```

```python
#explore dataset
df.head()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
#choose features and target 
X = df.drop('median_house_value', axis=1)#features. so we pick all columns except house_value
y = df['median_house_value']
```

```python
#split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.3, #30% of dataset shld be for testing. 70/30 split
    random_state = 89 #seeding value
) 
"""Note:
X_train - features columns for training. y_train - Target column for training
X_test - features columns for testing. y_test - Target column for testing
"""
```

```python
X_train
```

```python
X_test
```

```python
#create model
model = LinearRegression()
```

```python
#train the model
model.fit(X_train, y_train)
```

```python
#predict
prediction = model.predict(X_test)
```

```python
prediction.head()
```

```python
prediction
```

```python
#compare prediction
print("Actual: ", y_test.values)
print("Predicted: ", prediction)
```

```python
#evaluate
mse = mean_squared_error(y_test, prediction)
r2 = r2_score(y_test, prediction)
```

```python
r2
mse
```

```python
r2
```

```python
mse
```

```python
#put in a table
testCp = X_test.copy()
testCp["Actual Price"] = y_test
testCp["Predicted Price"] = prediction
testCp["Error"] = testCp["Actual Price"] - testCp["Predicted Price"]
```

```python
testCp.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, Lasso
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Ridge Regression (L2 regularization)
ridge = Ridge(alpha=1.0)#create model
ridge.fit(X_train_scaled, y_train)#train it
y_pred_ridge = ridge.predict(X_test_scaled)#predict
```

```python
#compare prediction
print("Actual: ", y_test.values)
print("Predicted: ", y_pred_ridge)
```

```python
#evaluate prediction
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
```

```python
mse_ridge
```

```python
r2_ridge
```

```python
#put in a table
testCp_Ridge = X_test.copy()
testCp_Ridge["Actual Price"] = y_test
testCp_Ridge["Predicted Price"] = y_pred_ridge
testCp_Ridge["Error"] = testCp_Ridge["Actual Price"] - testCp_Ridge["Predicted Price"]
```

```python
testCp_Ridge.head()
```

```python
# Lasso Regression (L1 regularization)
lasso = Lasso(alpha=0.01) #create model
lasso.fit(X_train_scaled, y_train) #train it
y_pred_lasso = lasso.predict(X_test_scaled) #predict
```

```python
#compare prediction
print("Actual: ", y_test.values)
print("Predicted: ", y_pred_lasso)
```

```python
#evaluate prediction
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)
```

```python
mse_lasso
```

```python
r2_lasso
```

```python
#put in a table
testCp_lasso = X_test.copy()
testCp_lasso["Actual Price"] = y_test
testCp_lasso["Predicted Price"] = y_pred_lasso
testCp_lasso["Error"] = testCp_lasso["Actual Price"] - testCp_lasso["Predicted Price"]
```

```python
testCp_lasso.head()
```

```python
#decide better model
print("R2 - Linear: ", r2)
print("R2 - Ridge: ", r2_ridge)
print("R2 - Lasso: ", r2_lasso)
print("-----")
print("Mse - Linear: ", mse)
print("Mse - Ridge: ", mse_ridge)
print("Mse - Lasso: ", mse_lasso)
```
